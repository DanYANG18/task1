```python
import pandas as pd
df = pd.read_excel('Desktop/NZ_Admin_JOBS.xlsx', header = None, skiprows = [0])
df.head()
```

```python
df.isnull().sum()
```

```python
df[0] = df[0].str.title()
df[0].head(10)
```

```python
df[3] = df[3].str.split(',', n=1, expand = True)[0]
df[3]
```

```python
df[['Location','Area']] = df[3].str.split('area: ', n=1, expand = True)
df
```

```python
df['Location'] = df['Location'].apply(lambda x:x[9:])
df['Location']
```

```python
def deduplication(x):
    trim = x.strip()
    index = (trim + trim).find(trim,1)
    if index != -1:
        return trim[:index]
    else:
        return trim
```

```python
df['Location'].isna().sum()
```

```python
df['Area'].isna().sum()
```

```python
df['Location'] = df['Location'].apply(deduplication)
df['Location']
```

```python
df['Area'].fillna(value = 'unknown', inplace = True)
df['Area'] = df['Area'].apply(deduplication)
df['Area']
```

```python
df.drop(3, axis = 1, inplace = True)
df.head()
```

```python
df[['Classification','sub-Classification']] = df[5].str.split('subClassification: ', n=1, expand = True)
df
```

```python
df.drop(columns = [5], inplace = True)
df
```

```python
df['Classification'] = df['Classification'].str.replace('classification: ','')
df
```

```python
df['Classification'].isna().sum()
```

```python
df['sub-Classification'].isna().sum()
```

```python
df['sub-Classification'].fillna(value = 'unknown', inplace = True)
df['sub-Classification'] = df['sub-Classification'].apply(deduplication)
df['Classification'] = df['Classification'].apply(deduplication)
df
```

```python
df[[4,8]] = df[4].str.split(',at',expand=True)
df
```

```python
df[2] = df[2].astype(str) + df[8].astype(str)
df
```

```python
df.drop(columns = [8], inplace = True)
df
```
